### 这是一个大模型学习笔记
step1: 运行命令将变更添加到缓存区，.表示将所有变更添加到缓存区，也可以替换成文件名--git add .<br>
step2：运行命令提交变更--git commit -m "变更的描述"<br>
step3: 运行命令推送变更到远程仓库--git push (origin main)<br>
